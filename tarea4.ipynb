### **Ejercicio 1**

Use el m√©todo de bisecci√≥n para encontrar soluciones precisas dentro de $10^{-2}$ para $ùë•^3 ‚àí 7ùë•^2 + 14ùë• ‚àí 6 = 0$ en cada intervalo. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n
def f(x):
    return x**3 - 7 * x**2 + 14 * x - 6

# Implementar el m√©todo de bisecci√≥n
def biseccion(func, a, b, tol=1e-2):
    if func(a) * func(b) >= 0:
        print("El m√©todo de bisecci√≥n no es aplicable en este intervalo.")
        return None
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if func(c) == 0:
            return c
        elif func(a) * func(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2
```

a. $[0, 1]$

```python
a, b = 0, 1
raiz_a = biseccion(f, a, b)

x_vals_a = np.linspace(a, b, 100)
y_vals_a = f(x_vals_a)

plt.figure(figsize=(8, 5))
plt.plot(x_vals_a, y_vals_a, label=r'$f(x) = x^3 - 7x^2 + 14x - 6$')
plt.axhline(0, color='black', linewidth=0.5)
if raiz_a is not None:
    plt.plot(raiz_a, f(raiz_a), 'ro', label=f'Ra√≠z: x ‚âà {raiz_a:.2f}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Soluci√≥n de la ecuaci√≥n en el intervalo [0, 1]')
plt.legend()
plt.grid(True)
plt.show()
```

b. Use el m√©todo de bisecci√≥n para encontrar una aproximaci√≥n dentro de $10^{-3}$ para un valor en $[‚àí2, 0]$ con $x^2 -1=e^{1-x^2}$.

```python
# Definir la funci√≥n para la bisecci√≥n
def f(x):
    return x**2 - 1 - np.exp(1 - x**2)

# Implementar el m√©todo de bisecci√≥n
def biseccion(func, a, b, tol=1e-3):
    if func(a) * func(b) >= 0:
        print("El m√©todo de bisecci√≥n no es aplicable en este intervalo.")
        return None
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if func(c) == 0:
            return c
        elif func(a) * func(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

# b. Aplicar el m√©todo de bisecci√≥n en el intervalo [-2, 0]
a, b = -2, 0
raiz = biseccion(f, a, b)

# Mostrar el resultado
print(f"Una aproximaci√≥n de la soluci√≥n de x^2 - 1 = e^(1 - x^2) en el intervalo [{a}, {b}] es x ‚âà {raiz:.3f}")
```

### **Ejercicio 1**

Un abrevadero de longitud $L$ tiene una secci√≥n transversal en forma de semic√≠rculo con radio $r$. (Consulte la figura adjunta.) Cuando se llena con agua hasta una distancia $‚Ñé$ a partir de la parte superior, el volumen $ùëâ$ de 
agua es

![](ADJUNTOS\ej1Aplicados.png)

Suponga que $ùêø = 10 ùëêùëö, ùëü = 1 ùëêùëö$ y $ùëâ = 12.4 ùëêùëö^3$. Encuentre la profundidad del agua en el abrevadero dentro de $0.01 ùëêùëö$. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
L = 10  # longitud en cm
r = 1   # radio en cm
V_target = 12.4  # volumen en cm^3

# Funci√≥n para calcular el volumen de agua en el abrevadero
def volumen(h):
    return L * ((0.5 * np.pi - r**2) * (np.arcsin(h/r)) - (h * (r**2 - h**2)**(1/2)))

# M√©todo de bisecci√≥n para encontrar h
def bisection_method(V_target, h_low, h_high, tol):
    while (h_high - h_low) > tol:
        h_mid = (h_low + h_high) / 2
        V_mid = volumen(h_mid)

        if V_mid < V_target:
            h_low = h_mid  # Aumentar h
        else:
            h_high = h_mid  # Disminuir h

    return (h_low + h_high) / 2

# Valores iniciales
h_low = 0
h_high = r  # h no puede ser mayor que r
tolerance = 0.01

# Encontrar la profundidad h
h_solution = bisection_method(V_target, h_low, h_high, tolerance)
print(f"Profundidad del agua: {h_solution:.4f} cm")

# Graficar la relaci√≥n entre h y V
h_values = np.linspace(0, r, 100)
V_values = volumen(h_values)

plt.figure(figsize=(10, 6))
plt.plot(h_values, V_values, label='Volumen de agua V(h)', color='blue')
plt.axhline(V_target, color='red', linestyle='--', label='V = 12.4 cm¬≥')
plt.axvline(h_solution, color='green', linestyle='--', label=f'h = {h_solution:.2f} cm')
plt.title('Volumen de Agua en el Abrevadero')
plt.xlabel('Profundidad del Agua (cm)')
plt.ylabel('Volumen (cm¬≥)')
plt.legend()
plt.grid()
plt.show()
```

### **Ejercicio 2**

Un objeto que cae verticalmente a trav√©s del aire est√° sujeto a una resistencia viscosa, as√≠ como a la fuerza de gravedad. Suponga que un objeto con masa $m cae desde una altura $S_0$, y que la altura del objeto despu√©s de $t$ segundos es

$$ s(t) = S_0 - \frac{mg}{k}t + \frac{m^2g}{k^2}(1 - e^{-kt/m}) $$

donde $g = 9.81 m/s^2$ y $k$ representa el coeficiente de la resistencia del aire en $Ns/m$. Suponga $S_0 = 300m$, $m = 0.25 kg$, y $k = 0.1 Ns/m $. Encuentre, dentro de $0.01$ segundos, el tiempo exacto que tarda un cuarto de $kg$ en golpear el piso.


```python
import numpy as np

# Constantes
S0 = 300  # altura inicial en metros
m = 0.25  # masa en kg
g = 9.81  # aceleraci√≥n debido a la gravedad en m/s^2
k = 0.1   # coeficiente de resistencia en Ns/m

# Funci√≥n para calcular la altura en funci√≥n del tiempo t
def altura(t):
    return S0 - (m * g / k) * t + (m**2 * g / k**2) * (1 - np.exp(-k * t / m))

# M√©todo de bisecci√≥n para encontrar el tiempo en que el objeto toca el suelo
def bisection_method(S0, t_low, t_high, tol):
    while (t_high - t_low) > tol:
        t_mid = (t_low + t_high) / 2
        s_mid = altura(t_mid)

        if s_mid > 0:
            t_low = t_mid  # Aumentar el tiempo
        else:
            t_high = t_mid  # Disminuir el tiempo

    return (t_low + t_high) / 2

# Valores iniciales
t_low = 0
t_high = 100  # Se establece un l√≠mite superior grande para garantizar que el objeto caiga
tolerance = 0.01

# Encontrar el tiempo de impacto
t_solution = bisection_method(S0, t_low, t_high, tolerance)
print(f"Tiempo que tarda en golpear el piso: {t_solution:.4f} segundos")
```

### **Ejercicio 1**

Use el teorema 2.1 para encontrar una cota para el n√∫mero de iteraciones necesarias para lograr una aproximaci√≥n con precisi√≥n de $10^-4$ para la soluci√≥n de $x^3-x-1=0$ que se encuentra dentro del intervalo $[1, 2]$ Encuentre una aproximaci√≥n para la ra√≠z con este grado de precisi√≥n.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funci√≥n
def f(x):
    return x**3 - x - 1

# Cota para el n√∫mero de iteraciones necesarias
def calculate_iterations(a, b, epsilon):
    return np.ceil(np.log2((b - a) / epsilon)).astype(int)

# M√©todo de bisecci√≥n
def bisection_method(a, b, epsilon):
    if f(a) * f(b) >= 0:
        raise ValueError("La funci√≥n debe tener diferentes signos en los extremos del intervalo.")

    iterations = calculate_iterations(a, b, epsilon)
    for _ in range(iterations):
        mid = (a + b) / 2
        if f(mid) == 0:
            return mid  # Se encontr√≥ la ra√≠z
        elif f(a) * f(mid) < 0:
            b = mid
        else:
            a = mid

    return (a + b) / 2  # Retorna la aproximaci√≥n final

# Definimos el intervalo y la precisi√≥n
a = 1
b = 2
epsilon = 1e-4



# Calculamos el n√∫mero de iteraciones necesarias
num_iterations = calculate_iterations(a, b, epsilon)
print(f"N√∫mero de iteraciones necesarias: {num_iterations}")

# Encontramos la ra√≠z
root_approximation = bisection_method(a, b, epsilon)
print(f"Aproximaci√≥n de la ra√≠z: {root_approximation:.6f}")

# Graficar la funci√≥n y la aproximaci√≥n de la ra√≠z
x_values = np.linspace(0.5, 2.5, 100)  # Rango de valores de x
y_values = f(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='$f(x) = x^3 - x - 1$', color='blue')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # L√≠nea y=0
plt.axvline(root_approximation, color='red', linestyle='--', label=f'Aproximaci√≥n de la ra√≠z: {root_approximation:.4f}')
plt.title('M√©todo de Bisecci√≥n para $f(x) = x^3 - x - 1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(0.5, 2.5)
plt.ylim(-2, 2)
plt.legend()
plt.grid()
plt.show()

#
```
