### **Ejercicio 1**

Use el método de bisección para encontrar soluciones precisas dentro de $10^{-2}$ para $𝑥^3 − 7𝑥^2 + 14𝑥 − 6 = 0$ en cada intervalo. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return x**3 - 7 * x**2 + 14 * x - 6

# Implementar el método de bisección
def biseccion(func, a, b, tol=1e-2):
    if func(a) * func(b) >= 0:
        print("El método de bisección no es aplicable en este intervalo.")
        return None
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if func(c) == 0:
            return c
        elif func(a) * func(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2
```

a. $[0, 1]$

```python
a, b = 0, 1
raiz_a = biseccion(f, a, b)

x_vals_a = np.linspace(a, b, 100)
y_vals_a = f(x_vals_a)

plt.figure(figsize=(8, 5))
plt.plot(x_vals_a, y_vals_a, label=r'$f(x) = x^3 - 7x^2 + 14x - 6$')
plt.axhline(0, color='black', linewidth=0.5)
if raiz_a is not None:
    plt.plot(raiz_a, f(raiz_a), 'ro', label=f'Raíz: x ≈ {raiz_a:.2f}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Solución de la ecuación en el intervalo [0, 1]')
plt.legend()
plt.grid(True)
plt.show()
```

b. $[1, 3.2]$


```python
a, b = 1, 3.2
raiz_b = biseccion(f, a, b)

x_vals_b = np.linspace(a, b, 100)
y_vals_b = f(x_vals_b)

plt.figure(figsize=(8, 5))
plt.plot(x_vals_b, y_vals_b, label=r'$f(x) = x^3 - 7x^2 + 14x - 6$')
plt.axhline(0, color='black', linewidth=0.5)
if raiz_b is not None:
    plt.plot(raiz_b, f(raiz_b), 'ro', label=f'Raíz: x ≈ {raiz_b:.2f}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Solución de la ecuación en el intervalo [1, 3.2]')
plt.legend()
plt.grid(True)
plt.show()
```

### **Ejercicio 1**

Un abrevadero de longitud $L$ tiene una sección transversal en forma de semicírculo con radio $r$. (Consulte la figura adjunta.) Cuando se llena con agua hasta una distancia $ℎ$ a partir de la parte superior, el volumen $𝑉$ de 
agua es

![](ADJUNTOS\ej1Aplicados.png)

Suponga que $𝐿 = 10 𝑐𝑚, 𝑟 = 1 𝑐𝑚$ y $𝑉 = 12.4 𝑐𝑚^3$. Encuentre la profundidad del agua en el abrevadero dentro de $0.01 𝑐𝑚$. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
L = 10  # longitud en cm
r = 1   # radio en cm
V_target = 12.4  # volumen en cm^3

# Función para calcular el volumen de agua en el abrevadero
def volumen(h):
    return L * ((0.5 * np.pi - r**2) * (np.arcsin(h/r)) - (h * (r**2 - h**2)**(1/2)))

# Método de bisección para encontrar h
def bisection_method(V_target, h_low, h_high, tol):
    while (h_high - h_low) > tol:
        h_mid = (h_low + h_high) / 2
        V_mid = volumen(h_mid)

        if V_mid < V_target:
            h_low = h_mid  # Aumentar h
        else:
            h_high = h_mid  # Disminuir h

    return (h_low + h_high) / 2

# Valores iniciales
h_low = 0
h_high = r  # h no puede ser mayor que r
tolerance = 0.01

# Encontrar la profundidad h
h_solution = bisection_method(V_target, h_low, h_high, tolerance)
print(f"Profundidad del agua: {h_solution:.4f} cm")

# Graficar la relación entre h y V
h_values = np.linspace(0, r, 100)
V_values = volumen(h_values)

plt.figure(figsize=(10, 6))
plt.plot(h_values, V_values, label='Volumen de agua V(h)', color='blue')
plt.axhline(V_target, color='red', linestyle='--', label='V = 12.4 cm³')
plt.axvline(h_solution, color='green', linestyle='--', label=f'h = {h_solution:.2f} cm')
plt.title('Volumen de Agua en el Abrevadero')
plt.xlabel('Profundidad del Agua (cm)')
plt.ylabel('Volumen (cm³)')
plt.legend()
plt.grid()
plt.show()
```

### **Ejercicio 2**

Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa $m cae desde una altura $S_0$, y que la altura del objeto después de $t$ segundos es

$$ s(t) = S_0 - \frac{mg}{k}t + \frac{m^2g}{k^2}(1 - e^{-kt/m}) $$

donde $g = 9.81 m/s^2$ y $k$ representa el coeficiente de la resistencia del aire en $Ns/m$. Suponga $S_0 = 300m$, $m = 0.25 kg$, y $k = 0.1 Ns/m $. Encuentre, dentro de $0.01$ segundos, el tiempo exacto que tarda un cuarto de $kg$ en golpear el piso.


```python
import numpy as np

# Constantes
S0 = 300  # altura inicial en metros
m = 0.25  # masa en kg
g = 9.81  # aceleración debido a la gravedad en m/s^2
k = 0.1   # coeficiente de resistencia en Ns/m

# Función para calcular la altura en función del tiempo t
def altura(t):
    return S0 - (m * g / k) * t + (m**2 * g / k**2) * (1 - np.exp(-k * t / m))

# Método de bisección para encontrar el tiempo en que el objeto toca el suelo
def bisection_method(S0, t_low, t_high, tol):
    while (t_high - t_low) > tol:
        t_mid = (t_low + t_high) / 2
        s_mid = altura(t_mid)

        if s_mid > 0:
            t_low = t_mid  # Aumentar el tiempo
        else:
            t_high = t_mid  # Disminuir el tiempo

    return (t_low + t_high) / 2

# Valores iniciales
t_low = 0
t_high = 100  # Se establece un límite superior grande para garantizar que el objeto caiga
tolerance = 0.01

# Encontrar el tiempo de impacto
t_solution = bisection_method(S0, t_low, t_high, tolerance)
print(f"Tiempo que tarda en golpear el piso: {t_solution:.4f} segundos")
```

### **Ejercicio 1**

Use el teorema 2.1 para encontrar una cota para el número de iteraciones necesarias para lograr una aproximación con precisión de $10^-4$ para la solución de $x^3-x-1=0$ que se encuentra dentro del intervalo $[1, 2]$ Encuentre una aproximación para la raíz con este grado de precisión.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función
def f(x):
    return x**3 - x - 1

# Cota para el número de iteraciones necesarias
def calculate_iterations(a, b, epsilon):
    return np.ceil(np.log2((b - a) / epsilon)).astype(int)

# Método de bisección
def bisection_method(a, b, epsilon):
    if f(a) * f(b) >= 0:
        raise ValueError("La función debe tener diferentes signos en los extremos del intervalo.")

    iterations = calculate_iterations(a, b, epsilon)
    for _ in range(iterations):
        mid = (a + b) / 2
        if f(mid) == 0:
            return mid  # Se encontró la raíz
        elif f(a) * f(mid) < 0:
            b = mid
        else:
            a = mid

    return (a + b) / 2  # Retorna la aproximación final

# Definimos el intervalo y la precisión
a = 1
b = 2
epsilon = 1e-4



# Calculamos el número de iteraciones necesarias
num_iterations = calculate_iterations(a, b, epsilon)
print(f"Número de iteraciones necesarias: {num_iterations}")

# Encontramos la raíz
root_approximation = bisection_method(a, b, epsilon)
print(f"Aproximación de la raíz: {root_approximation:.6f}")

# Graficar la función y la aproximación de la raíz
x_values = np.linspace(0.5, 2.5, 100)  # Rango de valores de x
y_values = f(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='$f(x) = x^3 - x - 1$', color='blue')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Línea y=0
plt.axvline(root_approximation, color='red', linestyle='--', label=f'Aproximación de la raíz: {root_approximation:.4f}')
plt.title('Método de Bisección para $f(x) = x^3 - x - 1$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(0.5, 2.5)
plt.ylim(-2, 2)
plt.legend()
plt.grid()
plt.show()

#
```
